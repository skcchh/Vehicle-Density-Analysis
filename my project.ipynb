```python

resized = False


def forFrame(frame_number, output_array, output_count, detected_frame):
    print("FOR FRAME " , frame_number)
    print("Output for each object : ", output_array)
    print("Output count for unique objects : ", output_count)
    
    plt.clf()
    
    this_colors = []
    labels = []
    sizes = []

    counter = 0

    for eachItem in output_count:
        counter += 1
        labels.append(eachItem + " = " + str(output_count[eachItem]))
        sizes.append(output_count[eachItem])
        this_colors.append(color_index[eachItem])
        
    plt.title("Analysis: " + str(frame_number))
    pie1 = plt.pie(sizes, labels=labels, colors=this_colors, shadow=True, startangle=140, autopct="%1.1f%%")
    # redraw the canvas
    fig.canvas.draw()

    # convert canvas to image
    img = np.fromstring(fig.canvas.tostring_rgb(), dtype=np.uint8,sep='')
    img  = img.reshape(fig.canvas.get_width_height()[::-1] + (3,))

    # img is rgb, convert to opencv's default bgr
    img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)


    # display image with opencv or any operation you like
    img = cv2.resize(img, (400, 300))
    cv2.imshow("plot",img)
    detected_frame = cv2.resize(detected_frame, (960, 600))
    cv2.imshow("detected_frame",detected_frame)
    cv2.waitKey(1)
    
detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , "yolo.h5"))
detector.loadModel()
detected_frame,video_path = detector.detectObjectsFromVideo(save_detected_video = False,
                                         input_file_path=os.path.join(execution_path, "traffic.mp4"),
                                         return_detected_frame = True,  per_frame_function=forFrame,
                                         frames_per_second=2, log_progress=True)


```
